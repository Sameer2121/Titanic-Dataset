```python
import matplotlib.pyplot as plt
```

 What are the survival rates across different passenger classes?

```python
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Plot survival rates by passenger class
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

How does age affect the likelihood of survival?

```python
age_survival_counts = df.groupby(['Age', 'Survived']).size().unstack(fill_value=0)

# Plot survival counts by age using a line plot
plt.figure(figsize=(10, 6))
age_survival_counts.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Survival Count by Age')
plt.legend(['Not Survived', 'Survived'])
plt.show()

```

What is the survival rate by gender?

```python
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()

# Plot survival rates by gender
sns.barplot(x=survival_rate_by_gender.index, y=survival_rate_by_gender.values)
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.show()
```

How do family size and survival rate relate?

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Plot the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Survived', y='FamilySize', data=df, palette='muted')
plt.xlabel('Survived')
plt.ylabel('Family Size')
plt.title('Distribution of Family Size by Survival')
plt.show()
```

What is the survival rate among passengers who boarded at different ports?

```python
embarked_counts = df['Embarked'].value_counts()
print(embarked_counts)
embarked_mapping = {1.0: 'C', 2.0: 'Q', 0.0: 'Q'}
df['Embarked'] = df['Embarked'].replace(embarked_mapping)

# Handle any remaining missing values in the 'Embarked' column
df['Embarked'] = df['Embarked'].fillna('S')  # Assuming 'S' for any remaining missing values

# Calculate survival rates by embarkation port
survival_rate_by_port = df.groupby('Embarked')['Survived'].mean().reset_index()
```

```python
# Calculate survival rates by embarkation port
survival_rate_by_port = df.groupby('Embarked')['Survived'].mean().reset_index()

# Plot survival rates by embarkation port
plt.figure(figsize=(10, 6))
sns.barplot(x='Embarked', y='Survived', data=survival_rate_by_port, order=['C', 'Q', 'S'])
plt.xlabel('Embarkation Port')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Embarkation Port')
plt.ylim(0, 1)  # Set the y-axis limits to range from 0 to 1 for clarity
plt.show()
```

How does fare price impact survival rate?

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Survived', y='Fare', data=df)
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.title('Distribution of Fare by Survival')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

Can we predict survival based on passenger demographics using a heatmap?

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()
```

```python

df['Embarked'] = df['Embarked'].fillna('S')

# Convert categorical variables to numeric
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])

# Create a correlation matrix
corr_matrix = df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Features')
plt.show()

```
